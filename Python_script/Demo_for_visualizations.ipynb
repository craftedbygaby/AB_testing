## Merging the Datasets

## Cleaning the Dataset

```python
info_com["group_test"].isna().sum()
```

```python
info_com.dropna(subset=["group_test"], inplace = True)
```

```python
info_com.head()
```

```python
info_com.isna().sum()
```

```python
info_com.dropna(thresh = 9).shape

# for the demographics, we'll need the most accurate data possible so we are dropping all the rows that aren't complete.
# which, by the way, are the same. all the null values are concentrated in 12 rows as you can see
# then there is a row with a null value in age. In that row we'll introduce the average.
```

```python
info_com.dropna(thresh = 9, inplace = True)
```

```python
# we fill the null value for age

info_com["age"] = info_com["age"].fillna(info_com["age"].mean())
```

## Columns

We'll create categories for age and we'll remove columns which aren't relevant for our visualizations.

```python
info_com.columns
```

```python
info_com.drop(columns = ['calls_last_6_months', 'logons_last_6_months'], inplace = True)
```

```python
def age_label(age):
    if age <18:
        return "Adolescent"
    elif age >= 18 and age <= 30:
        return "Young dults"
    elif age >= 31 and age < 60:
        return "Middle-aged"
    elif age >= 60 and age <= 90:
        return "Senior"
    else:
        return "Time to die"

info_com["age_label"] = info_com["age"].apply(age_label)
info_com.head()
```

```python
info_com.columns
```

```python
info_com = info_com[['client_id', 'group_test', 'age', 'age_label', 'gender',
                    'num_accts', 'balance', 'tenure_years', 'tenure_months']]

# reordering columns to keep it super clean
```

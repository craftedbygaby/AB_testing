# 03. Hypotesis testing - Hypotesis Number 2

### H0: test group time ≥ control group time
### H1: test group time < control group time

```python
for step in data['process_step'].unique():
    control = data[(data['group_test'] == 'control') & (data['process_step'] == step)]['time_spent'].dropna()
    test = data[(data['group_test'] == 'test') & (data['process_step'] == step)]['time_spent'].dropna()
    result = st.ttest_ind(control, test, alternative='less')
    print(f"{step} - p-value : {result.pvalue}")
```

Start: p-values are above 0.05 — we fail to reject H0. There is no significant evidence that the test group is faster on these steps.

Step 1: p-values are below 0.05 — we reject H0. The test group spends significantly less time on these steps than the control group.

Step 2: p-values are above 0.05 — we fail to reject H0. There is no significant evidence that the test group is faster on these steps.

Step 3: p-values are above 0.05 — we fail to reject H0. There is no significant evidence that the test group is faster on these steps.

H₀ (Null Hypothesis): 
The completion rate of the Test group is less than or equal to the completion rate of the Control group 
(p_test ≤ p_control).

H₁ (Alternative Hypothesis): 
The completion rate of the Test group is greater than the completion rate of the Control group 
(p_test > p_control).

Significance level: α = 0.05

```python
#setup
import pandas as pd
import numpy as np
from statsmodels.stats.proportion import proportions_ztest
from scipy.stats import ttest_ind

# load data
df = pd.read_csv("/Users/hinahaq/Downloads/data_file_project5/final_test_data.csv")
print("Loaded:", df.shape)
df.head()
```

```python
# Order matters: [Test, Control] for "larger" test
counts = [21731, 16046]  # Number of successes 
nobs = [33157, 30910] #Total Trials

stat, p_z = proportions_ztest(counts, nobs, alternative='larger')
print(f"Correct Z-test (Test > Control):")
print(f"Z = {stat:.3f}, p = {p_z:.6f}")
print("Reject H0:", "YES" if p_z < 0.05 else "NO")
```

**RESULT**
A one-sided proportions z-test showed that the Test group achieved a significantly higher completion rate than the Control group (65.5% vs. 51.9%; z = 35.04, p < 0.000001), indicating that the new design led to a statistically significant improvement in process completion.

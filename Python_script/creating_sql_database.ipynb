# Creating the SQL database

## Importing libraries

```python
import pandas as pd
import numpy as np
import pymysql
from sqlalchemy import create_engine # To create the engine and the connection
from sqlalchemy import text # To send queries to MySQL.
import getpass  # To get the password without showing the input
```

## Creating the engine

Use your MySQL's password to create the connection between Python and MySQL.

```python
password = getpass.getpass()

connection_string = 'mysql+pymysql://root:' + password + '@localhost/'
engine = create_engine(connection_string)
connection = engine.connect()
query = text("""CREATE DATABASE IF NOT EXISTS vanguard;""")
connection.execute(query)
```

## Creating and populating the tables in SQL

We already created the database. Now, we create a new connection, in this case, to connect with the database we just created.

```python
password = getpass.getpass()

bd = "vanguard"
connection_string = 'mysql+pymysql://root:' + password + '@localhost/'+bd
engine = create_engine(connection_string)
connection = engine.connect()
```

We need to create variables which store our csv files' location.

```python
client_info_loc = "../Data/Clean_data/df_final_demo_cleaned.csv"
client_info = pd.read_csv(client_info_loc)

final_data_test_loc = "../Data/Clean_data/final_test_data.csv"
final_data_test = pd.read_csv(final_data_test_loc)
```

```python
client_info.to_sql(name = "client_info", con = engine, schema = "vanguard", if_exists = "replace")
final_data_test.to_sql(name = "final_data_test", con = engine, schema = "vanguard", if_exists = "replace")
```

### Primary keys and foreign keys

With all the tables created and populated we need to assign primary keys and foreign keys.

```python
# Assign the primary keys
query = text("""ALTER TABLE `client_info`
                ADD PRIMARY KEY(`client_id`)""")
connection.execute(query)

# Altering table the tables to connect the foreign keys
query = text("""ALTER TABLE `final_data_test`
                ADD CONSTRAINT `fk_client_id`
                FOREIGN KEY(`client_id`) REFERENCES `client_info`(`client_id`)""")
connection.execute(query)
```
